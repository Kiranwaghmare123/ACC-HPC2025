```python
#Sets:Unordered collection od unique elements
#Creation:
set1={1,2,3,4,5}
set1
```

```python
#add() : addition
set1.add(6)
set1
```

```python
#remove()
set1.remove(3)
set1
```

```python
#Set operations
#Union: Combine operation
set1 = {1,2,3}
set2 = {11,22.33}
set3 = set1.union(set2)
set3
```

```python
#Intersection : common elemnts in set
set4 = set1 & set2
set4
```

```python
#Difference:
set5 = set1 - set2
set5
```

```python
#Symmetric Difference: Elements in either set but not both
set6 = set1 ^ set2
set6
```

```python
#Matplotlib:

import pandas as pd
import matplotlib.pyplot as plt
```

```python
X = [1,2,3,4]
y = [10,20,25,30]

plt.plot(X,y)
plt.title("Line Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
from matplotlib.lines import lineStyles
X = [1,2,3,4]
y = [10,20,25,30]

plt.plot(X,y, linestyle='--',marker='*', color='r')
plt.title("Line Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
X = [5,7,9,10,13]
y = [10,20,25,30,35]

plt.scatter(X,y, color='purple')
plt.title("Scatter Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
#Question: Import tips dataset.
# Design the following plots: Scatter plot, Line plot, Bar plot, Count plot, Box plot, Pair plot,
# Heatmap, Histogram, Jointplot
```

```python
import seaborn as sns
data = sns.load_dataset("tips")
data.head()
```

```python
sns.scatterplot(x="total_bill", y="tip", data=data)
```

```python
sns.lineplot(x="total_bill", y="tip", data=data)
```

```python
sns.barplot(x='day', y='total_bill', data=data)
```

```python
sns.countplot(  x='day', data=data)
```

```python
sns.boxplot(x='day', y='total_bill', data=data)
```

```python
sns.pairplot(data)
```

```python
numeric_data = data.select_dtypes(include=['number'])
corr = numeric_data.corr()
sns.heatmap(corr, annot=True)
```

```python
sns.jointplot(x='total_bill', y='tip', data=data)
```
